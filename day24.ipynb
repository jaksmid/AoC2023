```python
with open("day24.txt") as f:
    lines = f.readlines()
current_pattern = []
for i, line in enumerate(lines):
    if i != len(lines) -1:
            line = line[:-1]
    current_pattern.append(line)
current_pattern
```

```python
from dataclasses import dataclass


@dataclass
class Coordinate:
    x: int
    y: int
    z: int

@dataclass
class Velocity:
    x: int
    y: int
    z: int

class Hail():
    def __init__(self, initial_position, velocity) -> None:
        self.initial_position = initial_position
        self.velocity = velocity

    def __repr__(self) -> str:
        return str(self.initial_position) + " " + str(self.velocity)
```

```python
hails = []
for raw_hail in current_pattern:
    raw_nums = [int(x.strip()) for x in raw_hail.replace("@", ",").split(',')]
    hails.append(Hail(
        Coordinate(*raw_nums[:3]),
        Velocity(*raw_nums[3:])
    ))
hails
```

```python
def intersect(x0a,y0a,dxa,dya,x0b,y0b,dxb,dyb):
    if dxa*dyb-dxb*dya == 0:
        return None, None, None, None
    ta = (dyb*(x0b-x0a)-dxb*(y0b-y0a))/(dxa*dyb-dxb*dya)
    tb = (dya*(x0a-x0b)-dxa*(y0a-y0b))/(dxb*dya-dxa*dyb)
    return (x0a+dxa*ta,y0a+dya*ta, ta, tb)
intersect(19, 13, -2, 1, 18, 19,-1, -1)
```
