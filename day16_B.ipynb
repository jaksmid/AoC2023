```python
import numpy as np
```

```python
patterns = []
with open("day16.txt") as f:
    lines = f.readlines()
current_pattern = []
for line in lines:
    if len(line) > 3:
        line = line[:-1]
        current_pattern.append(line)
    else:
        patterns.append(current_pattern)
        current_pattern = []
patterns.append(current_pattern)
patterns[0]
#beware - new line must be at input
```

```python
pattern = patterns[0]
x_dim = len(pattern[0])
y_dim = len(pattern)
x_dim, y_dim
```

```python
symbols = np.zeros((y_dim, x_dim), dtype="str")
```

```python
numbers = []
for y, row in enumerate(pattern):
    for x, val in enumerate(row):
        symbols[y, x] = val
```

```python
def pretty_print(matrix):
    print(str(matrix).replace(' ','').replace("'","").replace('[','').replace(']',''))
```

```python
pretty_print(symbols)
```

```python
def gen_energized(starting):
    energized = np.zeros((y_dim, x_dim), dtype="int")
    explored = set()
    from collections import deque
    queue = deque()
    queue.append(starting)
    while len(queue):
        to_explore = queue.popleft()
        explored.add(to_explore)
        y, x, direction = to_explore
        if y < 0 or x < 0 or y >= y_dim or x>= x_dim:
            continue
        energized[y,x] = 1
        symbol = symbols[y, x]
        if symbol == '.' or (direction in ['R', 'L'] and symbol == '-') or  (direction in ['U', 'D'] and symbol == '|'):
            x_new = None
            y_new = None
            if direction == 'R':
                y_new = y
                x_new = x+1
            elif direction == 'L':
                y_new = y
                x_new = x-1
            if direction == 'U':
                y_new = y-1
                x_new = x
            elif direction == 'D':
                y_new = y+1
                x_new = x
            new_instruction = (y_new, x_new, direction)
            if new_instruction not in explored:
                queue.append(new_instruction)
        elif symbol == '/':
            x_new = None
            y_new = None
            new_dir = None
            if direction == 'R':
                y_new = y - 1
                x_new = x
                new_dir = 'U'
            elif direction == 'L':
                y_new = y +1
                x_new = x
                new_dir = 'D'
            if direction == 'U':
                y_new = y
                x_new = x +1
                new_dir = 'R'
            elif direction == 'D':
                y_new = y
                x_new = x - 1
                new_dir = 'L'
            new_instruction = (y_new, x_new, new_dir)
            if new_instruction not in explored:
                queue.append(new_instruction)
        elif symbol == '\\':
            x_new = None
            y_new = None
            new_dir = None
            if direction == 'R':
                y_new = y + 1
                x_new = x
                new_dir = 'D'
            elif direction == 'L':
                y_new = y - 1
                x_new = x
                new_dir = 'U'
            if direction == 'U':
                y_new = y
                x_new = x - 1
                new_dir = 'L'
            elif direction == 'D':
                y_new = y
                x_new = x + 1
                new_dir = 'R'
            new_instruction = (y_new, x_new, new_dir)
            if new_instruction not in explored:
                queue.append(new_instruction)
        elif symbol in ["-", "|"]:
            x_new = None
            y_new = None
            new_dir = None
            if direction == 'R':
                y_new = y + 1
                x_new = x
                new_dir = 'D'
            elif direction == 'L':
                y_new = y - 1
                x_new = x
                new_dir = 'U'
            if direction == 'U':
                y_new = y
                x_new = x - 1
                new_dir = 'L'
            elif direction == 'D':
                y_new = y
                x_new = x + 1
                new_dir = 'R'
            new_instruction = (y_new, x_new, new_dir)
            if new_instruction not in explored:
                queue.append(new_instruction)
            x_new = None
            y_new = None
            new_dir = None
            if direction == 'R':
                y_new = y - 1
                x_new = x
                new_dir = 'U'
            elif direction == 'L':
                y_new = y +1
                x_new = x
                new_dir = 'D'
            if direction == 'U':
                y_new = y
                x_new = x +1
                new_dir = 'R'
            elif direction == 'D':
                y_new = y
                x_new = x - 1
                new_dir = 'L'
            new_instruction = (y_new, x_new, new_dir)
            if new_instruction not in explored:
                queue.append(new_instruction)
    return energized.sum()
        


```

```python
results = []
for y in range(y_dim):
    results.append(gen_energized((y,0,'R')))
    results.append(gen_energized((y,x_dim-1,'L')))
for x in range(x_dim):
    results.append(gen_energized((0,x,'D')))
    results.append(gen_energized((y_dim-1,x,'U')))
max(results)
```
