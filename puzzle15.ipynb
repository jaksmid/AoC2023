```python
df = pd.read_csv("puzzle15.txt", header=None, sep=" = ")
df.columns = ["node", "next"]
```

```python
df
```

```python
next_both = df.next.str.replace("[()]", "", regex=True).str.split(", ")
df["left"] = next_both.str[0]
df["right"] = next_both.str[1]
df
```

```python
lefts = {}
rights = {}
for _, row in df.iterrows():
    name = row["node"]
    lefts[name] = row["left"]
    rights[name] = row["right"]
instructions_map = {
    "L": lefts,
    "R": rights
}
```

```python
#instructions_map
```

```python
instruction_nr = 0
current_node = "AAA"
while True:
    for instruction in instructions:
        instruction_nr += 1
        current_node = instructions_map[instruction][current_node]
        if current_node == "ZZZ":
            break
    if current_node == "ZZZ":
            break
instruction_nr
```
