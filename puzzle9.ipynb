```python
class ValRange():
    def __init__(self, destination, source, length) -> None:
        self.source = source
        self.destination = destination
        self.length = length

    def isinrange(self, value):
        if self.source <= value and value < self.source + self.length:
            return self.destination + (value - self.source)
```

```python
class Mapping():
    def __init__(self, ranges) -> None:
        self.ranges = ranges

    def do_map(self, value):
        for r in self.ranges:
            v = r.isinrange(value)
            if v:
                return v
        return value
    
    def __getitem__(self,index):
         return self.do_map(index)
```

```python
def parse_mapping(mapping):
    mapping_lines = mapping.split("\n")
    mapping_lines = mapping_lines[1:]
    mapping_lines = [[int(x) for x in mapping_line.split()] for mapping_line in mapping_lines]
    ranges = []
    for destination, source, length in mapping_lines:    
        ranges.append(ValRange(destination, source, length))
    return Mapping(ranges)
```

```python
seed_soil = parse_mapping(input[1])
soil_fertilizer = parse_mapping(input[2])
fertilizer_water = parse_mapping(input[3])
water_light = parse_mapping(input[4])
light_temperature = parse_mapping(input[5])
temperature_humidity = parse_mapping(input[6])
humidity_location = parse_mapping(input[7])
```

```python
def get_location(seed_nr):
    return humidity_location[temperature_humidity[light_temperature[water_light[fertilizer_water[soil_fertilizer[seed_soil[seed_nr]]]]]]]
```

```python
location = [ get_location(seed) for seed in seeds]
min(location)
```
