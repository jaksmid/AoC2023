```python
def calculate_score(ticket):
    winning, tip = ticket.split("|")
    winning = winning.strip().split(" ")
    winning = {int(x) for x in winning if x}
    tip = tip.strip().split(" ")
    #print(tip)
    tip = {int(x) for x in tip if x}
    intersection = winning.intersection(tip)
    if (len(intersection)):
        value = 2**(len(intersection) -1)
    else:
        value = 0
    return winning, tip, intersection, len(intersection), value
```

```python
calculate_score(" 41 48 83 86 17 | 83 86  6 31 17  9 48 53")
```

```python
df["value"] = df.numbers.apply(lambda x: calculate_score(x)[3])
df
```

```python
copies = [1] * (len(df))
```

```python
for i, val in enumerate(df["value"]):
    #print(val)
    current_card_copies = copies[i]
    if val:
        for winning_copy in range(i+1, i+1+ val):
            if winning_copy < len(copies):
                copies[winning_copy] = copies[winning_copy] + current_card_copies
```
