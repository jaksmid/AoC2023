```python
class LavaNode():
    def __init__(self, y, x, direction, charges, cost) -> None:
        self.x = x
        self.y = y
        self.direction = direction
        self.charges = charges
        self.cost = cost

    def __repr__(self) -> str:
        return f"{self.y} {self.x} total cost {self.cost}"
    
    def __lt__(self, other):
        if self.charges != self.charges:
            return self.charges < other.charges
        if self.x == other.x:
            return self.y < other.y
        return self.x < other.x
```
