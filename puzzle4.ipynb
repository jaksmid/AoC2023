```python
df = pd.read_csv("puzzle3.txt", header=None, sep=": ")
df.columns = ["game", "sequence"]
df
```

```python
df["game_id"] = df.game.str.split(" ").str[-1].astype("int32")
df
```

```python
def parse_sequence(sequence):
    shown_hands = sequence.replace("; ", ",").replace(", ", ",").split(",")
    values = {
        "red": [],
        "blue": [],
        "green": []
    }
    for hand in shown_hands:
        #print(hand)
        amount, color = hand.split(" ")
        amount = int(amount)
        values[color].append(amount)
    return values
```

```python
parse_sequence("3 red, 4 green; 5 red, 1 blue; 2 green")
```

```python
def get_power(sequence):
    try:
        values = parse_sequence(sequence)
    except:
        print(sequence)
    red_max = max(values["red"])
    blue_max = max(values["blue"])
    green_max = max(values["green"])
    return  red_max * blue_max * green_max, red_max, blue_max, green_max
    
```

```python
get_power("3 red, 4 green; 5 red, 1 blue; 2 green")
```

```python
get_power("3 red, 4 green; 53 red, 1 blue; 2 green")
```

```python
get_power("7 blue, 5 red; 10 red, 7 blue; 5 blue, 4 green, 15 red; 4 green, 6 red, 7 blue; 5 green, 8 blue, 4 red; 5 red, 4 blue, 3 green")
```

```python
df["power_raw"] = df.sequence.apply(get_power)
df["power"] = df.power_raw.str[0]
```
