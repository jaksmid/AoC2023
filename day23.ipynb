```python
with open("day23.txt") as f:
    lines = f.readlines()
current_pattern = []
for i, line in enumerate(lines):
    if i != len(lines) -1:
            line = line[:-1]
    if len(line) > 3:
        line = line[:-1]
        current_pattern.append(line)
current_pattern
```

```python
x_dim = len(current_pattern[0])
y_dim = len(current_pattern)
x_dim, y_dim
```

```python
import numpy as np
symbols = np.zeros((y_dim, x_dim), dtype="str")
```

```python
numbers = []
for y, row in enumerate(current_pattern):
    for x, val in enumerate(row):
        symbols[y, x] = val
```

```python
np.set_printoptions(edgeitems=30, linewidth=100000)
def pretty_print(matrix):
    print(str(matrix).replace(' ','').replace("'","").replace('[','').replace(']',''))
```

```python
pretty_print(symbols)
```

```python
def can_visit(destination, direction):
    y, x = destination
    y_max = symbols.shape[0]
    x_max = symbols.shape[1]
    if x < 0 or y < 0:
        return False, False
    if x >= x_max or y >= y_max:
        return False, False
    symbol = symbols[y, x]
    if  symbol == '.':
        return True, False
    elif symbol == '#':
        return False, False
    else:
        if direction == "R":
            return symbol == '>', True
        if direction == "L":
            return symbol == '<', True
        if direction == "U":
            return symbol == '^', True
        if direction == "D":
            return symbol == 'v', True
    raise Exception("Undefined")
```

```python
def get_longest_path(current_node, visited_nodes):
    y, x = current_node
    if y == y_dim - 1:
        return len(visited_nodes)
    path_lengths = []
    direction_up = y-1, x
    if direction_up not in visited_nodes:
        should_test, should_slide = can_visit(direction_up, "U")
        if should_test:
            visited_nodes_copy = visited_nodes.copy()
            visited_nodes_copy.add((y-1, x))
            if should_slide:
                direction_up = y-2, x
                visited_nodes_copy.add((y-2, x))
            path_lengths.append(get_longest_path(direction_up, visited_nodes_copy))
    direction_down = y+1, x
    if direction_down not in visited_nodes:
        should_test, should_slide = can_visit(direction_down, "D")
        if should_test:
            visited_nodes_copy = visited_nodes.copy()
            visited_nodes_copy.add((y+1, x))
            if should_slide:
                direction_down = y+2, x
                visited_nodes_copy.add((y+2, x))
            path_lengths.append(get_longest_path(direction_down, visited_nodes_copy))
    direction_left = y, x-1
    if direction_left not in visited_nodes:
        should_test, should_slide = can_visit(direction_left, "L")
        if should_test:
            visited_nodes_copy = visited_nodes.copy()
            visited_nodes_copy.add((y, x-1))
            if should_slide:
                direction_left = y, x-2
                visited_nodes_copy.add((y, x-2))
            path_lengths.append(get_longest_path(direction_left, visited_nodes_copy))
    direction_right = y, x+1
    if direction_right not in visited_nodes:
        should_test, should_slide = can_visit(direction_right, "R")
        if should_test:
            visited_nodes_copy = visited_nodes.copy()
            visited_nodes_copy.add((y, x+1))
            if should_slide:
                direction_right = y, x+2
                visited_nodes_copy.add((y, x+2))
            path_lengths.append(get_longest_path(direction_right, visited_nodes_copy))
    if len(path_lengths) == 0:
        return -1
    return max(path_lengths)
        

```
